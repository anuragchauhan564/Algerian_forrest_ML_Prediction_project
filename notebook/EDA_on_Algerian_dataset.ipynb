```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,'Region'] = 0
dataset.loc[122:,'Region'] = 1
```

```python
df = dataset
```

```python
df.head()
```

```python
df.info()
```

```python
# to con=vert the Region into int  type 
```

```python
df[['Region']] = df[['Region']].astype(int) 
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# removing the 122nd row
df = df.drop(122).reset_index(drop = True)
```

```python
df.columns
```

```python
# fix spaces in columns names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
# changing the requried columns aas integer data type
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
# changing the other columns to float data type
```

```python
objects = [fearurs for fearurs in df.columns if df[fearurs].dtype == 'O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# lets save the cleaned dataset 
df.to_csv("Algerioan_forest_firest_cleaned_dataset.csv",index=False)
```
